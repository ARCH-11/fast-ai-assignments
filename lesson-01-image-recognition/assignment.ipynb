## Import pyhton dependencies

```python
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

## Visualize some examples

```python
imgs = []
for i in range(5):
    imgs.append(mpimg.imread('./data/train-fast-ai/dogs/dog.' + str(i) + '.jpg'))
for i in range(5):
    imgs.append(mpimg.imread('./data/train-fast-ai/cats/cat.' + str(i) + '.jpg'))

plt.figure(figsize=(12,8))
for i in range(10):
    plt.subplot(5,5,i+1)
    plt.imshow(imgs[i])
    plt.axis('off')
plt.show()
plt.close()
```
